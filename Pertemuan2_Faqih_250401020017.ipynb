```python
import pandas as pd
import numpy as np

URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
# Output dari pd.read_csv adalah DataFrame
df = pd.read_csv(URL)
```

```python
# Berapa jumlah baris dan kolom dataset Titanic?
# Dapat dijawab menggunakan DataFrame.shape, output-nya (jumlah_baris, jumlah_kolom)
df_shape = df.shape
jumlah_baris = df_shape[0]
jumlah_kolom = df_shape[1]
print('Jumlah baris:', jumlah_baris)
print('Jumlah kolom:', jumlah_kolom)
```

```python
# Kolom mana yang memiliki nilai kosong (missing)?
# Dapat dijawab menggunakan DataFrame.info(), output-nya berupa informasi tentang jumlah entri (baris), jumlah kolom, nama-nama kolom, tipe data kolom, dan jumlah kolom yang terisi (Non-Null Count).
# Jika Non-Null Count kolom kurang dari jumlah baris, artinya kolom tersebut memiliki nilai kosong
print(df.info())

# Bisa dijawab juga menggunakan DataFrame.isnull().sum()
# DataFrame.isnull() mengembalikan DataFrame baru yang berisi indikator apakah nilai baris-kolom tersebut kosong atau tidak. True = kosong, False = tidak kosong
# Kemudian menggunakan DataFrame.sum(), yang mana akan menjumlahkan semua baris. Karena True = 1 dan False = 0. Maka hasilnya bisa digunakan untuk menghitung jumlah baris kosong dari tiap kolom
# Jika muncul 0, maka kolom tersebut tidak memiliki nilai kosong. Sebaliknya, maka memiliki nilai kosong
print(df.isnull().sum())
```

```python
# Berapa persen penumpang yang selamat (Survived=1)?
# Karena persentase jumlah selamat = (jumlah selamat / jumlah total) * 100
# Maka perlu memfilter DataFrame utama dengan kondisi Survived == 1, ini akan mengembalikan DataFrame baru yang hanya berisi penumpang yang selamat
df_survived = df[df['Survived'] == 1]
# Kemudian menggunakan DataFrame.shape yang memiliki output (jumlah_baris, jumlah_kolom) untuk mendapat jumlah baris yang selamat melalui index 0
survived_count = df_survived.shape[0]
# Lalu menggunakan DataFrame.shape pada DataFrame utama untuk mendapat jumlah baris total melalui index 0 juga
total_count = df.shape[0]
# Lalu dikalikan 100
print((survived_count / total_count) * 100)
```

```python
# Rata-rata usia penumpang wanita kelas 1 yang selamat?
# Tahapannya:
# 1. Buat DataFrame yang terfilter: Sex = female dan Pclass = 1 dan Survived = 1
df_female_class1_survived = df[(df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Survived'] == 1)]
# 2. Buat Series umur (Age) dari DataFrame yang sudah terfilter
df_female_class1_survived_age_series = df_female_class1_survived['Age']
# 3. Menggunakan Series.mean() untuk mendapatkan rata-rata dari Series umur
print(df_female_class1_survived_age_series.mean())
```

```python
# Kelas penumpang mana yang tingkat keselamatan tertinggi?
# Tahapannya:
# 1. Mengelompokkan DataFrame berdasarkan kelas (Pclass) menggunakan groupBy
grouped = df.groupby('Pclass')
# 2. Menggunakan .mean() pada 'Survived' untuk mendapatkan mean Survived dari tiap kelas.
# Ini akan mengembalikan sebuah Series dengan index = kelas penumpang, values = mean Survived tiap kelas
grouped_survived_mean = grouped['Survived'].mean()
# 3. Karena semakin tinggi mean Survived = tingkat keselamatan semakin tinggi
# Maka, untuk mendapatkan yang index (kelas penumpang) dengan value paling tinggi, kita gunakan Series.idxmax()
print(grouped_survived_mean.idxmax())
```
